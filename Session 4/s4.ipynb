# Full ML Workflow on a Messy Clinical Dataset

You have been provided with data from a clinical study investigating potential biomarkers of disease progression.

The dataset contains patient demographic information, laboratory measurements, and treatment data.

Your task is to:
- Preprocess the data as needed.
- Explore correlations and decide which features to keep.
- Choose and train an appropriate model.
- Evaluate your model using relevant metrics.
- Interpret which features are most predictive.

The data is in `clinical_study.csv`. Read it in a dataframe and implement the steps you consider necessary to model the data in the best way.

Enjoy!
